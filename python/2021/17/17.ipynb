```python
from collections import defaultdict


def visualize(initial_vel, target_x_range: range, target_y_range: range):
    """
    Visualizes the probe's path.

    >>> visualize((7, 2), range(20, 30), range(-10, -5))
    .............#....#............
    .......#..............#........
    ...............................
    S........................#.....
    ...............................
    ...............................
    ...........................#...
    ...............................
    ....................TTTTTTTTTTT
    ....................TTTTTTTTTTT
    ....................TTTTTTTT#TT
    ....................TTTTTTTTTTT
    ....................TTTTTTTTTTT
    ....................TTTTTTTTTTT
    """

    tx_range = range(target_x_range.start, target_x_range.stop + 1)
    ty_range = range(target_y_range.start, target_y_range.stop + 1)
    
    x, y = 0, 0
    vx, vy = initial_vel
    min_x, max_x = 0, tx_range.stop
    min_y, max_y = ty_range.start, 0

    positions = defaultdict(set)
    while x <= max_x and y >= min_y:
        x += vx
        y += vy

        vx -= vx // abs(vx) if vx else 0
        vy -= 1

        min_x = min(min_x, x)
        max_y = max(max_y, y)
        positions[y].add(x)
    
    for y in range(max_y, min_y - 1, -1):
        generator = (
            '#' if x in positions[y] else 
            'S' if x == 0 and y == 0 else 
            'T' if x in tx_range and y in ty_range else 
            '.' for x in range(min_x, max_x)
        )
        print(''.join(generator))
    
    

visualize((30, -5), range(20, 30), range(-10, -5))
# target area: x=124..174, y=-123..-86
# visualize((16, 30), range(124, 174), range(-123, -86))
```
